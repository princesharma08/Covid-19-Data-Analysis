```python
# Filter the dataframe
df1 = df[["continent", "location", "date", "total_cases", "total_deaths", "gdp_per_capita", "human_development_index"]]
df1.head()
```

```python
df.update(df1)
```

```python
# Data Cleaning
df.drop_duplicates()
missing_values = df.isnull().sum()
print("Missing values in all columns:", missing_values)
df.dropna(subset=["continent"])
df.fillna(0)
```

```python
# Date time format
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.month
df.head()
```

```python
# Data aggregation
df_groupby = df.groupby("continent").max()
df_groupby.head()
```

```python
# Feature engineering
df["total_deaths_to_total_cases"] = df["total_deaths"] / df["total_cases"]
df_groupby.head()
```

```python
# Data Visualization
# Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
plt.figure(figsize=(10, 6))
sns.distplot(df["gdp_per_capita"])
plt.show()
```

```python
# Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
plt.figure(figsize=(10, 6))
plt.scatter(df["total_cases"], df["gdp_per_capita"])
plt.xlabel("Total Cases")
plt.ylabel("GDP per Capita")
plt.show()
```

```python
# Plot a bar plot of 'continent' column with 'total_cases'.
plt.figure(figsize=(10, 6))
sns.catplot(x="continent", y="total_cases", data=df, kind="bar")
plt.show()
```

```python
# Save the df_groupby dataframe in your local drive using pandas.to_csv function
df_groupby.to_csv("df_groupby.csv")
```
